### plotly를 활용한 타이타닉 데이터 시각화

```python
import plotly
import cufflinks as cf
import seaborn as sns
import pandas as pd
```

```python
# 오프라인 모드에서도 인터렉티브한 그래픽을 가능하도록 하기
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
train = pd.read_csv("./titanic/train.csv")
# test = pd.read_csv("./titanic/test.csv")
```

```python
train.head()
```

```python
train.iplot(kind = "bar", x = "Pclass", y = "Survived")
```

```python
train["Age"].iplot(kind = "histogram", bins = 20, title = "Passenger's Age")
```

```python
train["Fare"].iplot(kind = "histogram", bins = 20, title = "Passenger's Age")
```

### 클래스별 요금 확인

```python
train['Fare_1'] = train[ train['Pclass']==1 ]['Fare']
train['Fare_2'] = train[ train['Pclass']==2 ]['Fare']
train['Fare_3'] = train[ train['Pclass']==3 ]['Fare']
```

```python
sel = ['Fare_1', 'Fare_2', 'Fare_3']
train[ sel ].iplot(kind='histogram', 
                   bins=30,
                   xTitle='Fare', 
                   yTitle='Count')
```
