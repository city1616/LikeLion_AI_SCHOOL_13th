### 데이터 수집
* xpath : //*[@id="country-table"]/div/div/table/tbody/tr/td[]
* td[2] : 국가정보
* td[3] : 확진자 정보
* Td[4] : 위중증
* td[5] : 사망자 정보
* td[6] : 완치
* td[7] : 치명(%)
* td[8] : 완치(%)
* td[9] : 발생률
* td[10] : 인구수

```python
all_data = []
for i in range(2, 11):
    tmp = '//*[@id="country-table"]/div/div/table/tbody/tr/td[{}]'.format(i)
    # print(tmp)
    sel_ele = driver.find_elements_by_xpath(tmp)
    
    
    column_data = []
    for one in sel_ele:
        # print(one.text)
        column_data.append(one.text)
    
    print( len(sel_ele) )
    all_data.append(column_data)
    print(column_data)
    print()
    
all_data
```

```python
for i in range(len(all_data)) :
    print(len(all_data[i]))
```

### 데이터 프레임 생성

```python
import pandas as pd

dict_dat = { "국가":all_data[0], 
             "확진자":all_data[1],
             "위중증":all_data[2],
             "사망자":all_data[3],
             "완치":all_data[4],
             "치명(%)":all_data[5],
             "완치(%)":all_data[6],
             "발생률":all_data[7],
             "인구수":all_data[8]
           }

corona_df = pd.DataFrame(dict_dat)
corona_df
```

### 데이터 전처리

```python
# 확진자 합계, 확진자 1일 컬럼 추가
# corona_df[["확진자_합계", "확진자_1일"]] = corona_df["확진자"].str.split("\n", expand = True)
corona_df["확진자_합계"] = corona_df["확진자"].str.split("\n").str[0]
corona_df["확진자_1일"] = corona_df["확진자"].str.split("\n").str[1]
corona_df.tail(3)
```

```python
# 사망자 합계, 사망자 1일 컬럼 추가
# corona_df[["사망자_합계", "사망자_1일"]] = corona_df["사망자"].str.split("\n", expand = True)
corona_df["사망자_합계"] = corona_df["사망자"].str.split("\n").str[0]
corona_df["사망자_1일"] = corona_df["사망자"].str.split("\n").str[1]
corona_df.head(3)
```

```python
# 완치 합계, 완치 1일 컬럼 추가
corona_df["완치_합계"] = corona_df["완치"].str.split("\n").str[0]
corona_df["완치_1일"] = corona_df["완치"].str.split("\n").str[1]
corona_df.head(3)
```
